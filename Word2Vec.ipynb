```python
from gensim.models import Word2Vec
sentences=[
    ["king","queen","prince","princess"],
    ["man","woman","boy","girl"],
    ["apple","banana","mango","grape"],
    ["car","bus","train","bicycle"],
]
model=Word2Vec(sentences,vector_size=10,window=5,min_count=1,sg=0)
similar_words=model.wv.most_similar("king")
print(similar_words)
```
