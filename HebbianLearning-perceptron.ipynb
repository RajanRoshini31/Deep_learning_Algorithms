```python
weights=[0,0]
bias=0
inputs=[[0,0],[0,1],[1,0],[1,1]]
outputs=[0,0,0,1]
#hebbian learning rule
def hebbian_learning(inputs,outputs,weights,bias):
    for i in range(len(inputs)):
      for j in range(len(weights)):
        weights[j]+=inputs[i][j]*outputs[i]
        bias+=outputs[i]
    return weights,bias

weights,bias=hebbian_learning(inputs,outputs,weights,bias)
print(" Final weights:",weights)
print(" Final bias:",bias)
```

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.unit_step_func(linear_output)

                update = self.lr * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def unit_step_func(self, x):
        return np.where(x >= 0, 1, 0)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.unit_step_func(linear_output)
        return y_predicted

# AND gate data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Train the perceptron
perceptron = Perceptron()
perceptron.fit(X, y)

# Test the perceptron
test_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
predictions = perceptron.predict(test_data)
print("Predictions for AND gate:", predictions)
```

```python
import numpy as np

# Define inputs and expected outputs
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Inputs (x1, x2)
T = np.array([0, 0, 0, 1])  # Target outputs

# Initialize weights and bias
w = np.array([0.0, 0.0])  # Weights (w1, w2)
b = 0.0  # Bias
learning_rate = 0.1  # Learning rate

# Activation function (step function)
def step_function(z):
    return 1 if z >= 0 else 0

# Training the perceptron
epochs = 11
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}")
    for i in range(len(X)):
        x = X[i]
        t = T[i]
        z = np.dot(w, x) + b  # Linear combination
        y = step_function(z)  # Predicted output
        error = t - y  # Calculate error

        # Weight and bias updates
        w += learning_rate * error * x
        b += learning_rate * error

        # Display computation
        print(f"x: {x}, t: {t}, z: {z:.1f}, y: {y}, error: {error}, w: {w}, b: {b:.1f}")
    print("-" * 50)

print("Training completed.")
```
