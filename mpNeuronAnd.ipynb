```python
def mp_neuron_and(x1,x2):
  w1,w2=1,1
  threhold=2
  if (w1*x1+w2*x2)>=threhold:
    return 1
  else:
    return 0
inputs =[(0,0),(0,1),(1,0),(1,1)]
print("MP Neuron and GATE Results")
for x1, x2 in inputs:
  outputs=mp_neuron_and(x1,x2)
  print(f"Input({x1},{x2}) Output: {outputs}")
```

```python
def mp_neuron_or(x1,x2):
  w1,w2=1,1
  threhold=1
  if (w1*x1+w2*x2)>=threhold:
    return 1
  else:
    return 0
inputs =[(0,0),(0,1),(1,0),(1,1)]
print("MP Neuron or GATE Results")
for x1, x2 in inputs:
  outputs=mp_neuron_or(x1,x2)
  print(f"Input({x1},{x2}) Output: {outputs}")
```

```python
def mp_neuron_not(x):
  w=-1
  threhold=0
  if (w*x)>=threhold:
    return 1
  else:
    return 0
inputs =[0,1]
print("MP Neuron NOT GATE Results")
for x in inputs:
  outputs=mp_neuron_not(x)
  print(f"Input({x}) Output: {outputs}")
```

```python
def mp_neuron_xor(x1, x2):
  # Implementing XOR using AND, OR, and NOT gates
  temp1 = mp_neuron_and(x1, x2)
  temp2 = mp_neuron_or(x1, x2)
  xor_output = mp_neuron_and(mp_neuron_not(temp1), temp2)
  return xor_output

inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
print("MP Neuron XOR Gate Results")
for x1, x2 in inputs:
  output = mp_neuron_xor(x1, x2)
  print(f"Input({x1},{x2}) Output: {output}")
```
