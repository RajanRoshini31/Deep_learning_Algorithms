```python
import random

def create_grid(size):
    grid = [[' ' for _ in range(size)] for _ in range(size)]
    return grid

def place_objects(grid, objects):
    size = len(grid)
    for obj in objects:
        while True:
            x, y = random.randint(0, size - 1), random.randint(0, size - 1)
            if grid[x][y] == ' ':
                grid[x][y] = obj
                break

def display_grid(grid):
    for row in grid:
        print(' | '.join(row))
        print('-' * (len(row) * 4 - 1))

# Initialize game
size = 4
grid = create_grid(size)
place_objects(grid, ['W', 'P', 'G'])  # Wumpus, Pit, Gold
player_position = [0, 0]

# Example game display
grid[player_position[0]][player_position[1]] = 'P'
display_grid(grid)
```

```python
import pygame
import math

# Initialize Pygame
pygame.init()

# Define colors
BLACK = (0, 0, 0)
LEMON_CHIFFON = (250, 250, 205)

# Screen dimensions
width = 800
height = 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption('Rose Curve in Python !')
screen.fill(LEMON_CHIFFON)


# Function to draw a Maurer Rose
def drawMaurerRose(n, d, size):
    points = []
    for i in range(361):
        k = i * d
        r = size * math.sin(math.radians(n * k))
        x = r * math.cos(math.radians(k))
        y = r * math.sin(math.radians(k))
        points.append((width / 2 + x, height / 2 + y))
    pygame.draw.lines(screen, BLACK, False, points, 5)


def drawPattern():
    drawMaurerRose(6, 79, 200)


# Initial drawing
drawPattern()
pygame.display.flip()


# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    pygame.display.update()

# Quit Pygame
pygame.quit()
```
