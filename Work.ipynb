**MNIST DATASET  ACCURACY :0.96**

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.datasets import mnist
import numpy as np

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize the data

# Fibonacci function to define batch sizes
def fibonacci(n):
    fib = [1, 2]  # Start with 1 and 2 (since 0 images wouldn't make sense)
    for i in range(2, n):
        next_fib = fib[i-1] + fib[i-2]
        if next_fib > len(x_train):  # Stop if Fibonacci number exceeds dataset size
            break
        fib.append(next_fib)
    return fib

# Get Fibonacci sequence for batch sizes
fib_batches = fibonacci(20)  # Generate Fibonacci-based batch sizes

# Define the model
model = Sequential([
    Input(shape=(28, 28)),  # Define input shape
    Flatten(),
    Dense(128, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model in Fibonacci-sized increments
for batch_size in fib_batches:
    print(f"\nTraining on {batch_size} samples...\n")

    model.fit(x_train[:batch_size], y_train[:batch_size], epochs=5, verbose=1)

# Evaluate final model performance
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\nFinal Test Accuracy: {test_acc:.2f}")
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential, mixed_precision
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.preprocessing import image_dataset_from_directory
import numpy as np
import os

# Enable mixed precision training for lower memory usage
mixed_precision.set_global_policy('mixed_float16')

# Set image size and path to dataset
img_size = (64, 64)  # Reduced from 128x128 to lower memory usage
data_path = "/content/drive/MyDrive/archive (1)/data"  # Root folder containing 'Blight', 'Common_Rust', etc.

# Load dataset
train_dataset = image_dataset_from_directory(
    data_path,
    validation_split=0.2,
    subset="training",
    seed=42,
    image_size=img_size,
    batch_size=None  # We'll manually handle batch sizes
)

val_dataset = image_dataset_from_directory(
    data_path,
    validation_split=0.2,
    subset="validation",
    seed=42,
    image_size=img_size,
    batch_size=32  # Use a fixed batch size for validation
)

# Normalize the dataset
normalization_layer = tf.keras.layers.Rescaling(1./255)
train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y))
val_dataset = val_dataset.map(lambda x, y: (normalization_layer(x), y))

# Prefetch data for better efficiency
train_dataset = train_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
val_dataset = val_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)

# Function to generate Fibonacci batch sizes within memory limits
def fibonacci(n, max_size):
    fib = [1, 2]
    for i in range(2, n):
        next_fib = fib[i-1] + fib[i-2]
        if next_fib > max_size // 2:  # Limit batch size growth
            break
        fib.append(next_fib)
    return fib

# Get batch sizes based on dataset size
max_size = sum(1 for _ in train_dataset)  # Number of training batches
fib_batches = fibonacci(20, max_size)

# Define the model
model = Sequential([
    Input(shape=(64, 64, 3)),  # Adjusted input shape for RGB images
    Flatten(),
    Dense(128, activation='relu'),
    Dense(128, activation='relu'),
    Dense(4, activation='softmax')  # 4 classes: Blight, Common_Rust, etc.
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model with Fibonacci batch sizes
for batch_size in fib_batches:
    print(f"\nTraining on batch size: {batch_size}\n")

    # Rebatch dataset dynamically
    train_dataset_batch = train_dataset.batch(batch_size)

    model.fit(train_dataset_batch, epochs=5, validation_data=val_dataset, verbose=1)

# Evaluate final model performance
test_loss, test_acc = model.evaluate(val_dataset)
print(f"\nFinal Validation Accuracy: {test_acc:.2f}")
```
